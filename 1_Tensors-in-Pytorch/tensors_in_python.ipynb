## Reduction Operations

```python
e = torch.randint(size=(2,3), low=0, high=10)
e
```

```python
torch.randint(0,10,(2,3))
```

```python
torch.sum(e)
```

```python
print(torch.sum(e, dim=1)) #rows
print(torch.sum(e, dim=0)) #colns
```

```python
#mean (only works with float or complex not long)
e = e.to(torch.float32)
print(e)
print(torch.mean(e))


#mean along col
print(torch.mean(e, dim=0))

#mean along row
print(torch.mean(e, dim=1))

#max
print(torch.max(e))

#min
print(torch.min(e))

#prod
print(torch.prod(e))

#std
print(torch.std(e))

#var
print(torch.var(e))


#argmax
print(torch.argmax(e))

#argmin
print(torch.argmin(e))
```

## Tensors Operations on GPU

```python
import torch
torch.cuda.is_available()
```

```python
device = torch.device('cuda')
device
```

```python
#creating a new tensor on GPU (gpu ram -> vram)
torch.rand((2,3), device=device)
```

```python
# or move an existing cpu tensor to gpu
a = torch.rand(2,3)
print(a)
```

```python
b = a.to(device)
```

```python
b + 5 # on gpu
```

```python
# compare gpu and cpu time
import torch
import time

size = 10000

matrix_cpu1 = torch.rand(size, size)
matrix_cpu2 = torch.rand(size, size)

matrix_gpu1 = matrix_cpu1.to(device)
matrix_gpu2 = matrix_cpu2.to(device)

start = time.time()
result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)
end = time.time()
print("cpu time: ", end-start)

start = time.time()
result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2)
end = time.time()
print("gpu time: ", end-start)
```
