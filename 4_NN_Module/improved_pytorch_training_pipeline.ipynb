```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
```

```python
#train-test-split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0], test_size=0.2)
```

```python
#scaling
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

## Numpy Array to Pytorch Tensor

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

## Define the model

```python
import torch.nn as nn
class MySimpleNN(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features,1),
        nn.Sigmoid()
    )

    # self.weights = torch.rand(X.shape[1],1, dtype=torch.float64, requires_grad=True)
    # self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X):
    out = self.network(X)
    return out
    # z = torch.matmul(X, self.weights) + self.bias
    # y_pred = torch.sigmoid(z)
    # return y_pred

  def loss_function(self, y_pred, y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)
    return -(y*torch.log(y_pred) + (1-y)*torch.log(1-y_pred)).mean()
```

## important parameters

```python
learning_rate = 0.1
epochs = 25
```

## Training Pipeline

```python
model = MySimpleNN(X_train_tensor.shape[1])
model = model.double() # Cast model parameters to float64
# model.weights, model.bias
# print(model.weights.shape)
# print(model.bias.shape)

#Define loop
for epoch in range(100):
  #forward pass
  y_pred = model(X_train_tensor)
  # y_pred = model.forward(X_train_tensor)
  # print(y_pred)

  #loss cal
  loss = model.loss_function(y_pred, y_train_tensor.double()) # Cast y_train_tensor to float64 for compatibility
  # print(loss)


  #backward
  loss.backward()

  #update params
  with torch.no_grad():
    for param in model.parameters():
      param -= learning_rate * param.grad
    # model.weights -= learning_rate * model.weights.grad
    # model.bias -= learning_rate * model.bias.grad

  #zero grad
  for param in model.parameters():
    param.grad.zero_()
  # model.weights.grad.zero_()
  # model.bias.grad.zero_()

  print(f'Epoch : {epoch+1}, Loss: {loss.item()}')
```

## Evaluation

```python
with torch.no_grad():
  y_pred = model(X_test_tensor)
  # y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.5).float()

  accuracy = (y_pred==y_test_tensor).float().mean()

print(f"Accuracy : {accuracy.item()}")
# print(y_pred, y_train_tensor)
```

# Using Linear Layer instead Of Sequential + builtIn gradient + builtIn loss fun

```python
class MySimpleNN(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, X):
    out = self.linear(X)
    out = self.sigmoid(out)
    return out

  # def loss_function(self, y_pred, y):
  #   epsilon = 1e-7
  #   y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)
  #   return -(y*torch.log(y_pred) + (1-y)*torch.log(1-y_pred)).mean()

  def accuracy(self, y_pred, y):
    y_pred = (y_pred > 0.5).float()
    accuracy = (y_pred==y).float().mean()
    return accuracy
```

```python
learning_rate = 0.1
epochs = 25
```

```python
loss_function = nn.BCELoss()
# type(loss_function)
```

```python
model = MySimpleNN(X_train_tensor.shape[1]).double()

optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)

#Define loop
for epoch in range(100):
  #forward pass
  y_pred = model(X_train_tensor)

  #loss cal
  # loss = model.loss_function(y_pred, y_train_tensor)
  loss = loss_function(y_pred, y_train_tensor.view(-1,1).double())

  #zero grad (new)
  optimizer.zero_grad()

  #backward
  loss.backward()

  #update params
  # with torch.no_grad():
  #   model.linear.weight -= learning_rate * model.linear.weight.grad
  #   model.linear.bias -= learning_rate * model.linear.bias.grad

  optimizer.step()

  #zero grad
  #do befoe backward
  # model.linear.weight.grad.zero_()
  # model.linear.bias.grad.zero_()

  print(f'Epoch : {epoch+1}, Loss: {loss.item()}')
```

```python
with torch.no_grad():
  y_pred = model(X_test_tensor)
  y_pred = (y_pred > 0.5).float()

  accuracy = (y_pred==y_test_tensor).float().mean()

print(f"Accuracy : {accuracy.item()}")
# print(y_pred, y_train_tensor)
```
