## Manually Neural Network Code

cgpa|placement
6.7 | 0

goal is to find dL/dW and dL/Db


single neuron- > z = wx + b
y_pred = 1/(1+e^-z) [sigmoid]

Loss = -[ytarget ln(ypred) + (1-ytarget)ln(1-ypred)]

```python
import torch

x = torch.tensor(6.7)
y = torch.tensor(0.0)

w = torch.tensor(1.0)
b = torch.tensor(0.0)
```

```python
def binary_cross_entropy(prediction, target):
  epsilon = 1e-8
  prediction = torch.clamp(prediction, epsilon, 1-epsilon)
  return -(target * torch.log(prediction) + (1-target) * torch.log(1-prediction))
```

```python
# Forward Pass

z = w * x + b
y_pred = torch.sigmoid(z)

loss = binary_cross_entropy(y_pred, y)
```

```python
#Derivatives
dloss_dy_pred = (y_pred - y) / (y_pred * (1-y_pred))
dy_pred_dz = y_pred * (1-y_pred)
dz_dw = x
dz_db = 1

dL_dW = dloss_dy_pred * dy_pred_dz * dz_dw
dL_dB = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f"manual gradient of loss wrt w: {dL_dW}")
print(f"manual gradient of loss wrt b: {dL_dB}")
```

## Autograd

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)

w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
z = w*x+b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
loss = binary_cross_entropy(y_pred, y)
loss
```

```python
loss.backward()
```
